```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
from scipy.optimize import curve_fit
```

```python
#Estudos de Casos 1
clima = pd.read_csv("climate_evaporation_dataset.csv")
clima
```

```python

plt.scatter(clima["temperature_c"], clima["evaporation_mm_day"])
plt.xlabel("Temperatura")
plt.ylabel("Evaporação")
plt.show()

#O grafico apresenta um aumento gradativo na evaporação a medida que a temperatura aumenta, aspecto físico esperado.
```

```python
# Substitua pelos seus dados experimentais
T = clima['temperature_c']
E = clima["evaporation_mm_day"]

# Modelo: E(T) = a * exp(bT)
def modelo(T, a, b):
    return a * np.exp(b * T)

# Ajuste não linear
parametros, cov = curve_fit(modelo, T, E)
a, b = parametros

print(f"a = {a}")
print(f"b = {b}")
```

```python
#Estudo de Casos 2
#Exemplo para Quantidade de demanda
market = pd.read_csv("coffee_market_dataset.csv")
market
```

```python
plt.scatter(market["price_usd"], market["quantity_demand"])
plt.xlabel("Preço")
plt.ylabel("Quantidade de Demanda")
plt.show()
```

```python
# Substitua pelos seus dados experimentais
T = market['price_usd']
E = market["quantity_demand"]

# Modelo: E(T) = a * exp(bT)
def modelo(T, a, b):
    return a - b*np.log(T)

# Ajuste não linear
parametros, cov = curve_fit(modelo, T, E)
a, b = parametros

print(f"a = {a}")
print(f"b = {b}")
```

```python
# Substitua pelos seus dados experimentais
T = market['price_usd']
E = market["quantity_supply"]

# Modelo: E(T) = a * exp(bT)
def modelo(T, c, d):
    return c * np.exp(d*T)

# Ajuste não linear
parametros, cov = curve_fit(modelo, T, E)
c, d = parametros

print(f"c = {c}")
print(f"d = {d}")
```
