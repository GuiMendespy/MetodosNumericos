```python
def sist_lin_tri_inf(A,b):

    n = len(b)
    x = np.empty(n)
    x[0] = b[0]/A[0, 0]
    for i in range(1,n):
        x[i] = (b[i] - np.sum(A[i,:i]*x[:i]))/A[i,i]
    return x

print(sist_lin_tri_inf(A,b))
```

```python
#Exercicio 1.4

def resolver_sistema_trivial(A, b):
    # Garantir que são arrays do numpy
    A = np.array(A, dtype=float)
    b = np.array(b, dtype=float)
    n = len(b)
    
    is_upper = np.allclose(A, np.triu(A))
    is_lower = np.allclose(A, np.tril(A))
    
    if is_upper and is_lower:
        print("Tipo identificado: Matriz Diagonal")
        # x = b_i / A_ii
        return b / np.diag(A)
        
    elif is_upper:
        print("Tipo identificado: Triangular Superior (Retro-substituição)")
        x = np.zeros(n)
        for i in range(n - 1, -1, -1):
            soma = np.dot(A[i, i+1:], x[i+1:])
            x[i] = (b[i] - soma) / A[i, i]
        return x
        
    elif is_lower:
        print("Tipo identificado: Triangular Inferior (Substituição Progressiva)")
        x = np.zeros(n)
        for i in range(n):
            soma = np.dot(A[i, :i], x[:i])
            x[i] = (b[i] - soma) / A[i, i]
        return x
    
    else:
        raise ValueError("A matriz não é diagonal, triangular superior ou inferior.")

# --- Teste com os dados do Exemplo 3.4 ---

A_sup = np.array([[3, 4, -5, 1], [0, 1, 1, -2], [0, 0, 4, -5], [0, 0, 0, 2]])
b_sup = np.array([-10, -1, 3, 2])

A_inf = np.array([[3, 0, 0, 0], [2, 1, 0, 0], [1, 0, 1, 0], [1, 1, 1, 1]])
b_inf = np.array([4, 2, 4, 2])

print(f"Solução Superior: {resolver_sistema_trivial(A_sup, b_sup)}")
print(f"Solução Inferior: {resolver_sistema_trivial(A_inf, b_inf)}")
```
