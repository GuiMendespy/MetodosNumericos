```python
#implementação do ponto fixo

#quetão 1 e questao 2

def g(x):
    return 2*np.sin(np.sqrt(x))

def ponto_fixo(x0, erro, max_iter):
    x_atual = x0

    for i in range(max_iter):
        x_prox = g(x_atual)

        if abs(x_prox - x_atual) < erro:
            print(f"Convergiu em {i+1} iteracoes")
            return x_prox

        x_atual = x_prox

    print("Nao convergiu")
    return x_atual

raiz = ponto_fixo(x0=0.5, erro=0.001, max_iter=100)
print("Raiz aproximada:", raiz)
```

```python
#questao 4

def funcaoA(x):
    return 2*x**4 + 4*x**3 + 3*x**2 - 10*x - 15
def funcaoB(x):
    return (x+3)*(x+1)*(x-2)**3
def funcaoC(x):
    return 5*x**3 + x**2 - np.exp(1-2*x) + np.cos(x) + 20
def funcaoD(x):
    return x*np.sin(x) + 4
def funcaoE(x):
    return (x-3)**5 * np.log(x)
def funcaoF(x):
    return (x**10) -1

def gA(x):
    return (2*x**4 + 4*x**3 + 3*x**2 - 15)/10
def gB(x):
    return 2 - ((x+3)*(x+1))**(1/3)
def gC(x):
    return ((np.exp(1-2*x) - np.cos(x) - x**2 - 20)/5)**(1/3)
def gD(x):
    return -4/np.sin(x)
def gE(x):
    return np.exp(-(x-3)**5)
def gF(x):
    return 1/(x**9)





def bissecao(a,b,limite,funcao):

    i = 0
    if(np.sign(funcao(a)) * np.sign(funcao(b)) != -1):
        return None
    elif(np.sign(funcao(a)) * np.sign(funcao(b)) == -1):
        while(1):
            x = (a + b)/2
            if(funcao(x) == 0):
                break
            b = x if np.sign(funcao(a)) * np.sign(funcao(x)) == -1 else b
            a = a if np.sign(funcao(a)) * np.sign(funcao(x)) != 1 else x
            i += 1
            if(i > limite):
                break
    return(x)

def falsa_posicao(xl,xu,limite,funcao):
    i = 0
    iteracao = 0
    x_old = 0

    while i < limite:

        x = xu - (funcao(xu)*(xl - xu))/(funcao(xl) - funcao(xu))
        iteracao += 1

        if iteracao > 1:
            i = abs((x - x_old)/x)*100

        if funcao(xl)*funcao(x) < 0:
            xu = x
        else:
            xl = x

        x_old = x

    return x

def ponto_fixo(x0, erro, max_iter, funcao):
    x_atual = x0

    for i in range(max_iter):
        x_prox = funcao(x_atual)

        if abs(x_prox - x_atual) < erro:
            print(f"Convergiu em {i+1} iteracoes")
            return x_prox

        x_atual = x_prox

    print("Nao convergiu no Ponto fixo")
    return x_atual

print("A) Metodo da bisseção: ", bissecao(0,3,200,funcaoA), "--Metodo da falsa Posição: ", falsa_posicao(0,3,20,funcaoA),"--","Metodo do ponto fixo: ", ponto_fixo(0,10**(-10),200,gA) )
print("B) Metodo da bisseção: ", bissecao(0,5,200,funcaoB), "--Metodo da falsa Posição: ", falsa_posicao(0,5,20,funcaoB),"--","Metodo do ponto fixo: ", ponto_fixo(0,10**(-10),200,gB) )
print("C) Metodo da bisseção: ", bissecao(-5,5,200,funcaoC), "--Metodo da falsa Posição: ", falsa_posicao(-5,5,20,funcaoC),"--","Metodo do ponto fixo: ", ponto_fixo(5,10**(-10),200,gC) )
print("D) Metodo da bisseção: ", bissecao(1,5,200,funcaoD), "--Metodo da falsa Posição: ", falsa_posicao(1,5,2,funcaoD),"--","Metodo do ponto fixo: ", ponto_fixo(1,10**(-10),200,gD) )
print("E) Metodo da bisseção: ", bissecao(2,5,200,funcaoE), "--Metodo da falsa Posição: ", falsa_posicao(2,5,1,funcaoE),"--","Metodo do ponto fixo: ", ponto_fixo(2,10**(-10),200,gE) )
print("F) Metodo da bisseção: ", bissecao(0.8,1.2,200,funcaoF), "--Metodo da falsa Posição: ", falsa_posicao(0.8,1.2,1,funcaoF),"--","Metodo do ponto fixo: ", ponto_fixo(1,10**(-10),200,gF) )

```
