```python
#Exercício 5.2 e 5.3

#Extrair uma submatriz de uma matriz original 

matriz = np.arange(100).reshape(10,10)

c1 = matriz[:5,:5]
c2 = matriz[5:,:5]
c3 = matriz[:5,5:]
c4 = matriz[5:,5:]
print(c1)
print(c2)
print(c3)
print(c4)

novaMatrizA = np.concatenate((c3,c4), axis = 0)
novaMatrizB = np.concatenate((c2,c1), axis = 0)
novaMatrizC = np.concatenate((novaMatrizA,novaMatrizB), axis = 1)

print(novaMatrizC)
```

```python
#Exercício 5.4

#Soma de matrizes de maneira simples

A = np.random.randint(0,10,(3,3))
B = np.random.randint(0,10,(3,3))

resultado = np.zeros((3,3))

for i in range(3):
    for j in range(3):
        resultado[i,j] = A[i,j] + B[i,j]

print(resultado)
```

```python
#Exercício 5.5  - Distibuitividade

A = np.random.rand(3,4)
B = np.random.rand(3,4)
sigma = np.random.rand()

expr1 = sigma*(A+B)
expr2 = sigma*A + sigma*B
expr3 = A*sigma + B*sigma

print(np.allclose(expr1, expr2))
print(np.allclose(expr1, expr3))
```

```python
#Exercício 5.6 - Multiplicação de matrizes

A = np.random.randint(0,10,(2,3))
B = np.random.randint(0,10,(3,2))

resultado = np.zeros((2,2))

for i in range(2):
    for j in range(2):
        for k in range(3):
            resultado[i,j] += A[i,k]*B[k,j]

print("Manual:")
print(resultado)

print("NumPy:")
print(A @ B)
```

```python
#Exercício 5.7 - Regra LIVE EVIL

L = np.random.rand(2,6)
I = np.random.rand(6,3)
V = np.random.rand(3,5)
E = np.random.rand(5,2)

prod = L @ I @ V @ E

# passo 2
res1 = prod.T

# passo 3 (errado)
#res2 = L.T @ I.T @ V.T @ E.T

# passo 4 (correto)
res3 = E.T @ V.T @ I.T @ L.T

#print(np.allclose(res1, res2)) # False
print(np.allclose(res1, res3)) # True
```

```python
#Exercício 5.8 - Matriz Assimétrica

def eh_simetrica(A):
    return np.allclose(A, A.T)

A = np.array([[1,2,3],
              [2,5,6],
              [3,6,9]])

print(eh_simetrica(A))
```

```python
#Exercício 5.9 - Criar matriz simétrica

A = np.random.rand(3,3)

simetrica = (A + A.T)/2

print(eh_simetrica(simetrica))
```

```python
#Exercício 5.10 - Matriz Vetor

A = np.random.rand(3,3)
s = np.random.rand(3)

resultado = A @ s

print(resultado)
```

```python
#Exercicio 5.11 - Matriz diagonal

ones = np.ones((4,4))

D = np.diag([1,4,9,16])
D_sqrt = np.diag([1,2,3,4])

print("Pré:")
print(D @ ones)

print("Pós:")
print(ones @ D)

print("Ambos:")
print(D_sqrt @ ones @ D_sqrt)
```

```python
#Exercicio 5.12 - Diagonal

D1 = np.diag([1,2,3])
D2 = np.diag([4,5,6])

mult_matriz = D1 @ D2
hadamard = D1 * D2

print(mult_matriz)
print(hadamard)
```
