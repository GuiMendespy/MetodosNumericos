```python
import sympy as sp
import numpy as np

def derivada_simbolica(f_expr):
    x = sp.Symbol('x')
    return sp.diff(f_expr, x)

def gerar_funcao_numerica(f_expr):
    x = sp.Symbol('x')
    return sp.lambdify(x, f_expr, 'numpy')

def newton_raphson(f_expr, x0, erro, max_iter):
    x = sp.Symbol('x')

    f_num = gerar_funcao_numerica(f_expr)
    df_expr = derivada_simbolica(f_expr)
    df_num = gerar_funcao_numerica(df_expr)

    x_atual = x0

    for i in range(max_iter):
        x_prox = x_atual - (f_num(x_atual) / df_num(x_atual))

        if abs(x_prox - x_atual) < erro:
            print(f"Convergiu em {i+1} iteracoes")
            return x_prox

        x_atual = x_prox

    print("Nao convergiu")
    return x_atual

x = sp.Symbol('x')
f = x ** 2

raiz = newton_raphson(f, x0= 2, erro=0.00001, max_iter=100)
print("Raiz aproximada:", raiz)
```

```python
import sympy as sp
import numpy as np

def gerar_funcao_numerica(f_expr):
    x = sp.Symbol('x')
    return sp.lambdify(x, f_expr, 'numpy')

def newton_raphson(f_expr, x0, x1, erro, max_iter):
    x = sp.Symbol('x')

    f_num = gerar_funcao_numerica(f_expr)

    x_anterior = x0
    x_atual = x1

    for i in range(max_iter):
        x_prox = x_atual - (f_num(x_atual) * (x_anterior - x_atual)/ (f_num(x_anterior) - f_num(x_atual) ))

        if abs(x_prox - x_atual) < erro:
            print(f"Convergiu em {i+1} iteracoes")
            return x_prox

        x_atual = x_prox

    print("Nao convergiu")
    return x_atual

x = sp.Symbol('x')
f = x ** 2

raiz = newton_raphson(f, x0=1 , x1 = 2, erro=0.0000000000001, max_iter=200)
print("Raiz aproximada:", raiz)
```

```python
import numpy as np

def funcaoA(x):
    return 2*x**4 + 4*x**3 + 3*x**2 - 10*x - 15

def funcaoB(x):
    return (x+3)*(x+1)*(x-2)**3

def funcaoC(x):
    return 5*x**3 + x**2 - np.exp(1-2*x) + np.cos(x) + 20

def funcaoD(x):
    return x*np.sin(x) + 4

def funcaoE(x):
    if x <= 0:
        return np.nan
    return (x-3)**5 * np.log(x)

def funcaoF(x):
    return (x**10) - 1


def gA(x):
    return (2*x**4 + 4*x**3 + 3*x**2 - 15)/10

def gB(x):
    return 2 - np.cbrt((x+3)*(x+1))

def gC(x):
    return np.cbrt((np.exp(1-2*x) - np.cos(x) - x**2 - 20)/5)

def gD(x):
    if np.sin(x) == 0:
        return np.nan
    return -4/np.sin(x)

def gE(x):
    return np.exp(-(x-3)**5)

def gF(x):
    if x == 0:
        return np.nan
    return 1/(x**9)


def dA(x):
    return 8*x**3 + 12*x**2 + 6*x - 10

def dB(x):
    return (x-2)**2 * (5*x**2 + 10*x - 1)

def dC(x):
    return 15*x**2 + 2*x + 2*np.exp(1 - 2*x) - np.sin(x)

def dD(x):
    return np.sin(x) + x*np.cos(x)

def dE(x):
    if x <= 0:
        return np.nan
    return 5*(x-3)**4 * np.log(x) + (x-3)**5 * (1/x)

def dF(x):
    return 10*x**9


# ================= MÉTODOS ================= #

def bissecao(a, b, limite, funcao):
    if np.sign(funcao(a)) * np.sign(funcao(b)) != -1:
        return None

    for _ in range(limite):
        x = (a + b) / 2

        if funcao(x) == 0:
            return x

        if np.sign(funcao(a)) * np.sign(funcao(x)) < 0:
            b = x
        else:
            a = x

    return x


def falsa_posicao(xl, xu, limite, funcao):
    x_old = 0

    for i in range(limite):
        denom = funcao(xl) - funcao(xu)
        if denom == 0:
            return None

        x = xu - (funcao(xu)*(xl - xu)) / denom

        if i > 0:
            erro = abs((x - x_old)/x)*100
            if erro < 1e-10:
                return x

        if funcao(xl)*funcao(x) < 0:
            xu = x
        else:
            xl = x

        x_old = x

    return x


def ponto_fixo(x0, erro, max_iter, funcao):
    x_atual = x0

    for i in range(max_iter):

        if abs(x_atual) > 1e6:
            print("Divergiu no Ponto Fixo")
            return None

        x_prox = funcao(x_atual)

        if np.isnan(x_prox) or np.isinf(x_prox):
            print("Erro numerico no Ponto Fixo")
            return None

        if abs(x_prox - x_atual) < erro:
            print(f"Convergiu em {i+1} iteracoes")
            return x_prox

        x_atual = x_prox

    print("Nao convergiu no Ponto Fixo")
    return x_atual


def newton_raphson(f, x0, erro, max_iter, df):
    x_atual = x0

    for i in range(max_iter):

        if df(x_atual) == 0:
            print("Derivada zero - falha no Newton")
            return None

        x_prox = x_atual - f(x_atual)/df(x_atual)

        if np.isnan(x_prox) or np.isinf(x_prox):
            print("Erro numerico no Newton")
            return None

        if abs(x_prox - x_atual) < erro:
            print(f"Convergiu em {i+1} iteracoes")
            return x_prox

        x_atual = x_prox

    print("Nao convergiu no Newton")
    return x_atual


def metodo_reta_secante(f, x0, x1, erro, max_iter):
    x_anterior = x0
    x_atual = x1

    for i in range(max_iter):

        denom = f(x_anterior) - f(x_atual)
        if denom == 0:
            print("Divisao por zero na Secante")
            return None

        x_prox = x_atual - f(x_atual)*(x_anterior - x_atual)/denom

        if np.isnan(x_prox) or np.isinf(x_prox):
            print("Erro numerico na Secante")
            return None

        if abs(x_prox - x_atual) < erro:
            print(f"Convergiu em {i+1} iteracoes")
            return x_prox

        x_anterior = x_atual
        x_atual = x_prox

    print("Nao convergiu na Secante")
    return x_atual


print("\nA)")
print("Bissecao:", bissecao(0,3,200,funcaoA))
print("Falsa Posicao:", falsa_posicao(0,3,200,funcaoA))
print("Ponto Fixo:", ponto_fixo(0,1e-10,200,gA))
print("Newton:", newton_raphson(funcaoA,1,1e-10,200,dA))
print("Secante:", metodo_reta_secante(funcaoA,0,3,1e-10,200))

print("\nB)")
print("Bissecao:", bissecao(0,5,200,funcaoB))
print("Falsa Posicao:", falsa_posicao(0,5,200,funcaoB))
print("Ponto Fixo:", ponto_fixo(0,1e-10,200,gB))
print("Newton:", newton_raphson(funcaoB,1,1e-10,200,dB))
print("Secante:", metodo_reta_secante(funcaoB,0,5,1e-10,200))
```
