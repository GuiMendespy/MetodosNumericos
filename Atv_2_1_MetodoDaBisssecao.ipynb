```python
#Exercício 3

def funcao(x):
    #f = np.sin(x) - np.cos(3*x)
    f = np.sin(x) - x**3
    return f


i = 0
a = 0.5
b = 1
erro = 100
iteracao = 0
while erro > 2:

    x = (a + b) / 2
    iteracao += 1

    if funcao(a) * funcao(x) < 0:
        b = x
    else:
        a = x

    erro = abs((b - a) / x) * 100

    print(f"Iteração {iteracao}: x = {x:.5f} | erro = {erro:.2f}%")

print("\nRaiz aproximada:", x)
```

```python
#Exercício 4

def funcao(x):
    return -2*x**6 - 1.5*x**4 + 10*x + 20

def derivada(x):
    return -12*x**5 - 6*x**3 + 10


a = 0
b = 1
erro = 100
iteracao = 0

while erro > 5:

    x = (a + b) / 2
    iteracao += 1

    if derivada(a) * derivada(x) < 0:
        b = x
    else:
        a = x

    erro = abs((b - a) / x) * 100

    print(f"Iteração {iteracao}: x = {x:.5f} | erro = {erro:.2f}%")

print("\nPonto crítico:", x)
print("Valor da função:", funcao(x))
```
