# Aula 1 - 🔢 Métodos de Agregação

Principais métodos para resumir dados:

.sum() → Soma dos elementos

.min() → Valor mínimo

.max() → Valor máximo

.mean() → Média dos valores

.cumsum() → Soma acumulada

📌 Argumento axis

Define a direção da agregação:

axis=0 → agrega por colunas

axis=1 → agrega por linhas


```python
import numpy as np

array = np.array([[1, 2, 3],
                  [4, 5, 6]])

array.sum(axis=0)  # Soma por coluna
array.sum(axis=1)  # Soma por linha
```

📌 Argumento keepdims

Mantém a dimensionalidade original do array:

```python
array.sum(axis=1, keepdims=True)
```

# Aula 2 - ⚡ Operações Vetorizadas

NumPy permite realizar operações diretamente sobre arrays sem usar loops.

➕ Operações com escalar

```python
array + 3
array * 3
```

➕ Operações entre arrays

```python
array_a = np.array([1,2,3,4,5,6,7,8])
array_b = np.array([9,10,11,12,13,14,15,16])

print(array_a + array_b)
print(array_a * 3)
```

🔍 Comparações Vetorizadas

Operações booleanas também funcionam diretamente:

```python
array > 2
```

🧠 Vetorizando funções Python

```python
array_d = np.array(["NumPy", "is", "awesome"])

vectorized_len = np.vectorize(len)
vectorized_len(array_d) > 2
#print(array_d) 
```

# Aula 03 - 📡 Broadcasting

Broadcasting permite operações entre arrays de tamanhos diferentes, seguindo regras específicas.

📏 Regras de Compatibilidade

NumPy compara dimensões da direita para a esquerda.

Duas dimensões são compatíveis quando:

São iguais, ou

Uma delas é 1

✅ Exemplos Compatíveis

(10, 5) e (10, 1)

(10, 5) e (5,)

❌ Exemplos Incompatíveis

(10, 5) e (5, 10)

(10, 5) e (10,)

🔄 Broadcasting de Linhas

```python
array = np.arange(10).reshape((2, 5))
print(array)
array + np.array([0, 1, 2, 3, 4])

# (2,5) e (1,5)
```

🔄 Broadcasting de Colunas

```python
array + np.array([0, 1]).reshape((2, 1))

# (2,5) e (2,1)
```
