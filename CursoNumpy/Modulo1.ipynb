# 📘 Aula 1 — Introdução ao NumPy

## 🔹 O que é NumPy?

NumPy é uma biblioteca fundamental para computação numérica em Python.  
Ela permite trabalhar com **arrays multidimensionais** de forma eficiente.

### ✅ Principais vantagens:
- Melhor uso de memória
- Operações matemáticas vetorizadas
- Alto desempenho
- Base para bibliotecas como Pandas, SciPy e Scikit-learn

Usa menos memória

Mais rápido para cálculos matemáticos

# 📘 Aula 2 — Dimensionalidade de Arrays
🔹 Dimensões

1D → Vetor

2D → Matriz

3D ou mais → Tensor

🔹 Shape

O atributo .shape mostra as dimensões do array.

````python
array = np.zeros((3,5))
print(array)
array.shape
````

🔹 Flatten

Transforma um array em 1D:

````python
val = array.flatten()
print(val)
print(array)
````

🔹 Reshape

Altera o formato mantendo o número de elementos.

````python
array.reshape((5,3))
````

🔹 Type Coercion (Conversão automática)

Quando misturamos tipos:

bool + int → vira int

int + float → vira float

misturar com string → tudo vira string

````python
np.array([True, "Boop", 42])
````
