````python
sudoku_game = np.array([
    [0, 0, 4, 3, 0, 0, 2, 0, 9],
    [0, 0, 5, 0, 0, 9, 0, 0, 1],
    [0, 7, 0, 0, 6, 0, 0, 4, 3],
    [0, 0, 6, 0, 0, 2, 0, 8, 7],
    [1, 9, 0, 0, 0, 7, 4, 0, 0],
    [0, 5, 0, 0, 8, 3, 0, 0, 0],
    [6, 0, 0, 0, 0, 0, 1, 0, 5],
    [0, 0, 3, 5, 0, 8, 6, 9, 0],
    [0, 4, 2, 9, 1, 0, 3, 0, 0]
])
````

````python
sudoku_game[2, 4]
````

Formato geral:

array[linha,coluna]

Linha inteira

````python
sudoku_game[0]
````

Coluna inteira


````python
sudoku_game[:, 3]
````

📘 Slicing (Fatiamento)


🔹 Slicing 1D

````python
array = [1,2,3,4,5,6,7]
````

````python
array[2:4]
````

Formato geral:


array[inicio:fim]

🔹 Slicing 2D

````python
sudoku_game[3:6, 3:6]
````

Formato:

array[linhas,colunas]

🔹 Slicing com Step

````python
sudoku_game[3:6:2, 3:6:2]
````

Formato:

array[inicio:fim:passo]

📘 Ordenação (Sorting)

🔹 Ordenação padrão

````python
np.sort(sudoku_game)
````

🔹 Ordenando por eixo

````python
np.sort(sudoku_game, axis=0)
````

📌 Conceito importante:

axis=0 → ordena por colunas

axis=1 → ordena por linhas

# Aula 2 - 📘 Filtering Arrays

🔹 Duas formas de filtrar:

Boolean Masks + Fancy Indexing

np.where()

📘 Boolean Masks


````python
one_to_five = np.arange(1, 6)
mask = one_to_five % 2 == 0
````

Mask gerada: 

[False,True,False,True,False]

🔹 Fancy Indexing (1D)

````python
one_to_five[mask]
````

Resultado:

[2,4]

🔹 Fancy Indexing (2D)

````python
classroom_ids_and_sizes = np.array([[1, 22], [2, 21], [3, 27], [4, 26]])
print(classroom_ids_and_sizes)
classroom_ids_and_sizes[:, 0][classroom_ids_and_sizes[:, 1] % 2 == 0]
````

📘 np.where()

🔹 Retorna índices

````python
np.where(classroom_ids_and_sizes[:, 1] % 2 == 0)
````

Retorna:

(array([0,3]),)

🔹 Recuperando linhas e colunas

````python
row_ind, column_ind = np.where(sudoku_game == 0)

````

📌 np.where() retorna uma tupla de índices.


🔹 Find and Replace

````python
np.where(sudoku_game == 0, "", sudoku_game)

````

Formato geral:

np.where(condicao,valor_se_verdadeiro,valor_se_falso)

# Aula 3 - Adicionando e Removendo Dados

🔹 Concatenando Linhas

````python
array1= np.array([[1, 22], [2, 21], [3, 27], [4, 26]])
array2= np.array([[1,"James"], [1,"George"], [3,"Amy"],[3,"Meehir"]])
np.concatenate((array1, array2))
````

Por padrão:

axis=0


🔹 Concatenando Colunas

````python
np.concatenate((array1, array2), axis=1)
````

📘 Compatibilidade de Shape

Se um array tem formato:

(
𝑚
,
𝑛
)
(m,n)

Para concatenar:

Mesmo número de colunas → axis=0

Mesmo número de linhas → axis=1

🔹 Criando compatibilidade com reshape

````python
array1.reshape((8,1))
array2.reshape((8,1))
````

📘 Removendo Dados

🔹 Remover linha

````python
np.delete(array1, 1, axis=0)
````

🔹 Remover coluna

````python
np.delete(array1, 1, axis=1)
````

🔹 Remover sem especificar eixo

````python
np.delete(array1, 1)
````

⚠ Sem eixo → o array é achatado (flatten).
