# Aula 02 - 🖼 Arrays RGB (Imagens)

Imagens podem ser representadas como arrays 3D:

(altura, largura, canais_de_cor)

Os canais são:

0 → 🔴 Vermelho (Red)

1 → 🟢 Verde (Green)

2 → 🔵 Azul (Blue)

Exemplo de exibição:

```python
rgb_array = np.array([
    [[255, 0, 0],   [255, 0, 0],   [255, 0, 0]],    # Vermelho
    [[0, 255, 0],   [0, 255, 0],   [0, 255, 0]],    # Verde
    [[0, 0, 255],   [0, 0, 255],   [0, 0, 255]]     # Azul
])

plt.imshow(rgb_array)
plt.show()
```

🎨 Separando Canais RGB

```python
red_array = rgb_array[:, :, 0]
green_array = rgb_array[:, :, 1]
blue_array = rgb_array[:, :, 2]

print("Red:\n", red_array)
print("Green:\n", green_array)
print("Blue:\n", blue_array)
```

🌑 Modificando Valores com np.where

```python
dark_logo_array = np.where(rgb_array == 255, 50, rgb_array)

plt.imshow(dark_logo_array)
plt.title("Logo Escurecida")
plt.show()
```

💾 Salvando como .npy

```python
with open("dark_logo.npy", "wb") as f:
    np.save(f, dark_logo_array)
```

📥 Carregando o Arquivo

```python
with open("dark_logo.npy", "rb") as f:
    loaded_array = np.load(f)

plt.imshow(loaded_array)
plt.title("Logo Carregada")
plt.show()
```

🔄 Flip (Espelhamento)

```python
flipped_logo = np.flip(rgb_array)

plt.imshow(flipped_logo)
plt.title("Flip Completo")
plt.show()
```

Flip por eixo

```python
flip_rows = np.flip(rgb_array, axis=0)
flip_columns = np.flip(rgb_array, axis=1)
flip_colors = np.flip(rgb_array, axis=2)

plt.imshow(flip_rows)
plt.title("Flip axis 0")
plt.show()

plt.imshow(flip_columns)
plt.title("Flip axis 1")
plt.show()

plt.imshow(flip_colors)
plt.title("Flip axis 2")
plt.show()
```

🔁 Transpose (Transposição)

```python
transposed_logo = np.transpose(rgb_array, axes=(1, 0, 2))

plt.imshow(transposed_logo)
plt.title("Transposto")
plt.show()
```

✂️ Split (Dividindo os Canais)

```python
red_split, green_split, blue_split = np.split(rgb_array, 3, axis=2)

print("Shape Red Split:", red_split.shape)
```

Removendo dimensão extra:

```python
red_2D = red_split.reshape((3, 3))
print(red_2D)
```

🧱 Stack (Empilhando Novamente)

```python
stacked_rgb = np.stack([red_2D, green_array, blue_array], axis=2)

plt.imshow(stacked_rgb)
plt.title("Stack Reconstruído")
plt.show()
```
